#### Data Preprocessing

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
result = pd.read_csv("result.csv")
del result['Unnamed: 0']
```

```python
print result.shape
result.columns
```

##### Split outcome variables

```python
outcome = result.iloc[:,11:]
outcome = pd.concat([result[['imdb_id']], outcome], axis=1)
outcome = outcome.set_index("imdb_id")
print outcome.shape
outcome.head()
```

##### Bag-of-words analysis for title and plot, get words which has a some weight

```python
result_copy = result.copy()
result_copy['title_plot'] = result_copy['title'].astype(str) + ' ' + result_copy['overview'].astype(str)
result_copy = result_copy[['movie_id','imdb_id', 'title_plot']]
```

```python
print result_copy.shape
result_copy.head()
```

```python
vectorizer = TfidfVectorizer(stop_words='english', min_df=0.02, encoding = "ISO-8859-1")
values = result_copy['title_plot'].values
vectorized = vectorizer.fit_transform(values)
words = vectorizer.get_feature_names()
len(words)
```

```python
vectorized_array = vectorized.toarray()
```

```python
bagofwords = pd.DataFrame(vectorized_array, columns=words)
bagofwords = bagofwords.add_prefix('word_')
bagofwords = pd.concat([result[['imdb_id']], bagofwords], axis=1)
bagofwords = bagofwords.set_index("imdb_id")
print bagofwords.shape
bagofwords.head()
```

##### Split features

```python
features = result.iloc[:,:11]
del features["overview"]
del features["title"]
del features["movie_id"]
features = features.set_index("imdb_id")
features.shape
```

##### Merge features and bagofword values

```python
merged = pd.concat([features, bagofwords], axis=1)
print merged.shape
merged.head()
```

##### One-hot-encoding for 'year' feature

```python
merged = merged.dropna( subset = ["year"] )
```

```python
df_with_dummies = pd.get_dummies( merged, columns = ["year"] )
```

```python
print df_with_dummies.shape
df_with_dummies.head()
```

```python
columns_data = df_with_dummies.columns
data = df_with_dummies.dropna()
data.head()
```

##### Scale all values with StandardScaler

```python
scaler = StandardScaler()
scaled_values_ndarray = scaler.fit_transform(data)
scaled_df = pd.DataFrame(scaled_values_ndarray, columns=columns_data, index=data.index)
```

```python
scaled_df.head()
```

```python
final_df = scaled_df
final_df['imdb_id'] = final_df.index
final_df.shape
```

```python
final_outcome = outcome
final_outcome['imdb_id'] = outcome.index
print final_outcome.shape
final_outcome.head()
```

```python
temp = pd.merge(final_df, outcome, how='inner', on=["imdb_id"])
temp.head()
```

```python
outcomes = temp.iloc[:,228:]
features = temp.iloc[:,:228]
del features["imdb_id"]
print outcomes.shape
print features.shape
```
